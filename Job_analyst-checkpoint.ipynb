```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Đọc dữ liệu từ các nguồn (giả sử dữ liệu đã được lưu dưới dạng CSV)
df_topcv = pd.read_csv('data_clean.topcv.csv')
df_jobgo = pd.read_csv('data_clean.jobgo.csv')
df_vietnamwork = pd.read_csv('data_clean.vietnamwork.csv')

# Thêm cột để đánh dấu nguồn dữ liệu
df_topcv['source'] = 'TopCV'
df_jobgo['source'] = 'Jobgo'
df_vietnamwork['source'] = 'Vietnamwork'

# Gộp cả 3 bảng lại thành 1 dataframe
df_all = pd.concat([df_topcv, df_jobgo, df_vietnamwork], ignore_index=True)
```

```python
# Tổng số công việc từ mỗi nguồn
total_jobs = df_all.groupby('source').size()

# Vẽ biểu đồ tổng số công việc
plt.figure(figsize=(8, 6))
total_jobs.plot(kind='bar', color=['blue', 'orange', 'green'])
plt.title('Tổng số công việc từ 3 nguồn')
plt.xlabel('Nguồn')
plt.ylabel('Số lượng công việc')
plt.show()
```

```python
# Tỷ lệ công việc giữa 3 nguồn
plt.figure(figsize=(8, 6))
df_all['source'].value_counts().plot(kind='pie', autopct='%1.1f%%', colors=['blue', 'orange', 'green'])
plt.title('Tỷ lệ các công việc giữa 3 nguồn')
plt.ylabel('')  # Ẩn nhãn trục Y
plt.show()
```

```python
industry_counts = df_all['industry'].value_counts()
print(industry_counts.values)
# Vẽ biểu đồ số lượng công việc theo ngành nghề
plt.figure(figsize=(10, 6))
industry_counts.plot(kind='bar', color='skyblue')
plt.title('Tỷ lệ các công việc theo ngành nghề')
plt.xlabel('Ngành nghề')
plt.ylabel('Số lượng công việc')
plt.xticks(rotation=90)
plt.show()
```

```python
# Đếm số lượng công việc theo hình thức làm việc và ngành nghề
job_type_counts = df_all.groupby(['industry', 'job_type']).size().unstack()

# Vẽ biểu đồ phân bố các công việc theo hình thức làm việc
plt.figure(figsize=(10, 6))
job_type_counts.plot(kind='bar', stacked=True, colormap='coolwarm')
plt.title('Phân bố các công việc theo hình thức làm việc ở các ngành nghề')
plt.xlabel('Ngành nghề')
plt.ylabel('Số lượng công việc')
plt.xticks(rotation=90)
plt.legend(title='Hình thức làm việc')
plt.show()
```

```python
# Đếm số lượng công việc yêu cầu giới tính theo ngành nghề
gender_counts = df_all.groupby(['industry', 'gender']).size().unstack()

# Vẽ biểu đồ yêu cầu về giới tính theo ngành nghề
plt.figure(figsize=(10, 6))
gender_counts.plot(kind='bar', stacked=True, colormap='coolwarm')
plt.title('Yêu cầu về giới tính trong các công việc ở các ngành nghề')
plt.xlabel('Ngành nghề')
plt.ylabel('Số lượng công việc')
plt.xticks(rotation=90)
plt.legend(title='Giới tính yêu cầu')
plt.show()
```

```python
df_with_education = df_all[df_all['min_education'].notna()]

# Đếm số lượng công việc theo ngành nghề và trình độ học vấn
education_counts = df_with_education.groupby(['industry', 'min_education']).size().unstack()

# Vẽ biểu đồ yêu cầu về trình độ học vấn theo ngành nghề
plt.figure(figsize=(10, 6))
education_counts.plot(kind='bar', stacked=True, colormap='coolwarm')
plt.title('Yêu cầu về trình độ học vấn ở các ngành nghề')
plt.xlabel('Ngành nghề')
plt.ylabel('Số lượng công việc')
plt.xticks(rotation=90)
plt.legend(title='Trình độ học vấn')
plt.show()
```

```python
# Tính mức lương trung bình của các công ty
average_salary = df_all.groupby('company')[['salary_min', 'salary_max']].mean().reset_index().head(30)
average_salary['average_salary'] = (average_salary['salary_min'] + average_salary['salary_max']) / 2

# Vẽ biểu đồ mức lương trung bình của các công ty
plt.figure(figsize=(10, 10))
average_salary[['company', 'average_salary']].set_index('company').plot(kind='bar', color='lightgreen')
plt.title('Mức lương trung bình của các công ty')
plt.xlabel('Công ty')
plt.ylabel('Mức lương trung bình')
plt.xticks(rotation=90)
plt.show()
```

```python
def classify_salary(row):
    if pd.isnull(row['salary_min']) and pd.isnull(row['salary_max']):
        return 'Thương lượng'
    elif pd.notnull(row['salary_min']) and pd.isnull(row['salary_max']):
        return 'Lương tối thiểu'
    elif pd.isnull(row['salary_min']) and pd.notnull(row['salary_max']):
        return 'Lương tối đa'
    elif pd.notnull(row['salary_min']) and pd.notnull(row['salary_max']):
        if row['salary_min'] == row['salary_max']:
            return 'Lương bằng giá trị'
        else:
            return 'Lương trong khoảng'
    return 'Không xác định'
```

```python
df_all['salary_category'] = df_all.apply(classify_salary, axis=1)
```

```python
salary_distribution = df_all['salary_category'].value_counts()
print(salary_distribution)
```

```python
df_all['average_salary'] = df_all[['salary_min', 'salary_max']].mean(axis=1)

average_salary_by_category = df_all.groupby('salary_category')['average_salary'].mean().reset_index()
print(average_salary_by_category)
```

```python
industry_company_counts = df_all.groupby(['industry', 'company']).size().reset_index(name='job_count')

# Tìm công ty lớn nhất trong mỗi ngành
top_companies = industry_company_counts.loc[industry_company_counts.groupby('industry')['job_count'].idxmax()]

# Tạo màu sắc cho mỗi công ty
colors = plt.cm.get_cmap('Set3', len(top_companies['company'].unique()))

# Vẽ biểu đồ
plt.figure(figsize=(12, 6))
bars = plt.bar(top_companies['industry'], top_companies['job_count'], color=colors(np.arange(len(top_companies))))

# Thêm tiêu đề và nhãn
plt.title('Các công ty lớn nhất theo ngành nghề')
plt.xlabel('Ngành nghề')
plt.ylabel('Số lượng tin tuyển dụng')
plt.xticks(rotation=90)

# Thêm chú thích cho tên công ty với màu sắc tương ứng
plt.legend([bars[i] for i in range(len(bars))], top_companies['company'], title='Công ty', loc='upper right')

plt.tight_layout()
plt.show()
```
